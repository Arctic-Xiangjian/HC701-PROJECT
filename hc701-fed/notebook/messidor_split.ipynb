```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from PIL import Image
from sklearn.model_selection import StratifiedShuffleSplit
```

## Pairs

```python
data_info = pd.read_csv('/ssd_Samsung870_2T/hc701-fed/hc701-project/hc701-fed_data/messidor/messidor_pairs/Info_pairs.csv')
```

```python
data_info
```

```python
valied_data = {'images' : [], 'labels' : []}
# image_id if has jpg needs be capitalized
for i,j in zip(data_info['Image name'],data_info['Retinopathy grade']):
    valied_data['images'].append(i)
    valied_data['labels'].append(j)
```

```python
image_path = '/ssd_Samsung870_2T/hc701-fed/hc701-project/hc701-fed_data/messidor/messidor_pairs/images'
X= valied_data['images']
y= valied_data['labels']
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.25, random_state=0)
for train_index, test_index in sss.split(X, y):
    train_set_index = train_index
    test_set_index = test_index
Train_data = {}
Test_data = {}
for i in train_set_index:
    image = Image.open(os.path.join(image_path,X[i])).convert('RGB')
    image = np.array(image)
    Train_data['image'] = image
    Train_data['label'] = y[i]
    np.save('/ssd_Samsung870_2T/hc701-fed/hc701-project/hc701-fed_data/messidor/messidor_pairs/train/'+X[i][:-4],Train_data)
for i in test_set_index:
    image = Image.open(os.path.join(image_path,X[i])).convert('RGB')
    image = np.array(image)
    Test_data['image'] = image
    Test_data['label'] = y[i]
    np.save('/ssd_Samsung870_2T/hc701-fed/hc701-project/hc701-fed_data/messidor/messidor_pairs/test/'+X[i][:-4],Test_data)
```
